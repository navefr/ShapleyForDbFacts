```python
import sys  
sys.path.insert(0, "../src")

from pathlib import Path
from circuit_shapley import CircuitShapley
import graphviz
```

```python
cid = "Ile7ci"
nnf_fname = Path("../circuits/Example/circuit_%s.nnf" % cid)
gates_fname = Path("../circuits/Example/circuit_%s_gate" %cid)
```

```python
c_shapley = CircuitShapley(nnf_fname, gates_fname)
```

```python
dot_fname = Path("../circuits/Example/circuit_%s.dot" % cid)

with open(dot_fname, "w") as f:
    f.write(c_shapley._ddnnf.to_dot())
with open(dot_fname, "r") as f:
    dot_graph = f.read()
graphviz.Source(dot_graph)    
```
